## Linear Quadratic Taylor Appximations

```python
import sympy as sp
```

```python
x = sp.Matrix(['x1','x2'])
```

```python
x
```

```python
f = x[0]**2 + x[1]**2
```

```python
f
```

```python
gf = sp.Matrix([f]).jacobian(x).T
```

```python
gf
```

```python
Hf = sp.Matrix([f]).jacobian(x).T.jacobian(x)
```

```python
Hf
```

#Numpy

```python
import numpy as np
```

```python
def f(x):
    return x[0]**2 + x[1]**2

def gf(x):
    return 2*x

def Hf(x):
    return np.array[[2,0],[0,2]]
```

```python
import matplotlib.pyplot as plt

x1 = np.linspace(-1,1,100)
x2 = np.linspace(-1,1,100)
X1,X2 = np.meshgrid(x1,x2)

F = X1**2 + X2**2

plt.contourf(X1,X2,F)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.colorbar()
plt.show()
```

```python
x0 = np.array([0.5,0.75])


x1 = np.linspace(-1,1,100)
x2 = np.linspace(-1,1,100)
X1,X2 = np.meshgrid(x1,x2)

F = X1**2 + X2**2

plt.contourf(X1,X2,F)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.plot(x0[0],x0[1],'ro', ms = 10)
plt.colorbar()
plt.show()
```

```python
x0 = np.array([0.5,0.75])

f0 = f(x0)

gf0 = gf(x0)
```

```python
f0, gf0
```

```python
L = np.zeros((100,100))
for i, x1 in enumerate(np.linspace(-1, 1, 100)):
  for j, x2 in enumerate(np.linspace(-1, 1, 100)):
    x = np.array([x1, x2])
    dx0 = x - x0
    L[i,j] = f0 + dx0.dot(gf0)

plt.contourf(X1,X2,L)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.plot(x0[0],x0[1],'ro', ms = 10)
plt.colorbar()
plt.show()
```

```python
Q = np.zeros((100,100))
for i, x1 in enumerate(np.linspace(-1, 1, 100)):
  for j, x2 in enumerate(np.linspace(-1, 1, 100)):
    x = np.array([x1, x2])
    dx0 = x - x0
    Q[i,j] = f0 + dx0.dot(gf0) + 0.5*dx0.dot(Hf(x0)).dot(dx0)

plt.contourf(X1,X2,L)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.plot(x0[0],x0[1],'ro', ms = 10)
plt.colorbar()
plt.show()
```
